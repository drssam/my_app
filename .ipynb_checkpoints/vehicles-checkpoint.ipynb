4. DATA CLEANING & Pre-processing

```python
# Alternative: fill missing values (example using forward fill)
df['is_4wd'] = df['is_4wd'].fillna(0).astype('Int16')
#df = df.fillna('np.nan')
df.isnull().sum()
```

```python
# Convert data types
df.info()
df.head()
```

```python
# Convert data types efficiently
df = df.astype({
    'price': 'float',
    'model_year': 'Int32',
    'cylinders': 'Int32',
    'odometer': 'Int32',
    'is_4wd': 'bool',
    'days_listed': 'Int16'
})

# Convert 'date_posted' to datetime format
df['date_posted'] = pd.to_datetime(df['date_posted'], errors='coerce')
```

```python
# 2.Logical Consistency
df = df[df['model_year'] <= 2025]
df['age'] = 2025 - df['model_year']
```

```python
df['manufacturer'] = df['model'].str.split().str[0]
df.head()
```

```python
def age_category(x):
    if x<5: return '<5'
    elif  x>=5 and x<10: return '5-10'
    elif x>=10 and x<20: return '10-20'
    else: return '>20'
df['age_category'] = df['age'].apply(age_category)
df.head()
```

```python
# 1. Detect Outliers (Example: Price)
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
#print(f"Number of outliers in price: {outliers.shape[0]}")
#df[['price']].sort_values(by='price', ascending=False).head(10)  # Check top 10 highest prices
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='type', y='price', showfliers=False, width=0.4)  # Exclude extreme outliers
price_ticks = np.arange(0, df['price'].max(), 5000)
plt.xticks(rotation=45)
plt.title('Price Distribution per Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Price ($)')
plt.show()
```

```python

import seaborn as sns
from matplotlib.widgets import Slider

# Set up the figure and initial plot
fig, ax = plt.subplots(figsize=(12,6))
plt.subplots_adjust(bottom=0.25)  # Make space for the slider

# Initial limits (show all prices)
min_price, max_price = df['price'].min(), df['price'].max()

# Draw initial boxplot
boxplot = sns.boxplot(data=df, x='type', y='price', showfliers=True, width=0.4)
plt.xticks(rotation=45)
plt.title('Interactive Price Distribution per Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Price ($)')

# Add a slider for adjusting the upper price limit
ax_slider = plt.axes([0.2, 0.1, 0.65, 0.03])  # Position of slider (left, bottom, width, height)
price_slider = Slider(ax_slider, 'Max Price', min_price, max_price, valinit=max_price, valstep=500)

# Update function for the slider
def update(val):
    ax.clear()  # Clear previous plot
    max_val = price_slider.val  # Get the current slider value

    # Redraw boxplot with new max price limit
    sns.boxplot(data=df[df['price'] <= max_val], x='type', y='price', showfliers=True, width=0.4, ax=ax)
    plt.xticks(rotation=45)
    plt.title('Interactive Price Distribution per Vehicle Type')
    plt.xlabel('Vehicle Type')
    plt.ylabel('Price ($)')
    fig.canvas.draw_idle()  # Update the plot

# Connect slider to update function
price_slider.on_changed(update)

plt.show()
```
